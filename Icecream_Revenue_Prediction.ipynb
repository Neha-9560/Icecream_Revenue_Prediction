```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create dataset
data = {
    'Temperature': [30, 32, 35, 36, 38, 40, 42, 45],
    'Revenue': [22, 25, 30, 32, 35, 40, 43, 48]
}
df = pd.DataFrame(data)

# Step 3: Visualize the data
plt.scatter(df['Temperature'], df['Revenue'], color='blue')
plt.title('Temperature vs Ice Cream Revenue')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue (₹1000s)')
plt.grid(True)
plt.show()

# Step 4: Prepare data
X = df[['Temperature']]  # Feature
y = df['Revenue']        # Target

# Step 5: Train the model
model = LinearRegression()
model.fit(X, y)

# Step 6: Predict revenue
predicted = model.predict(X)

# Step 7: Visualize regression line
plt.scatter(X, y, color='blue')
plt.plot(X, predicted, color='red')  # Regression line
plt.title('Linear Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue (₹1000s)')
plt.show()

# Step 8: Predict for a new temperature
new_temp = [[37]]
predicted_revenue = model.predict(new_temp)
print(f"Predicted revenue for {new_temp[0][0]}°C: ₹{predicted_revenue[0]*1000:.2f}")
```

### Project Summary:

This project uses a simple Linear Regression model to predict Ice Cream Revenue based on Temperature. As temperature increases, ice cream sales also increase. The dataset was small, but the model successfully learned a linear relationship and provided accurate revenue predictions for new temperatures.
